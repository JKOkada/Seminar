##Ⅱ．データの加工
###Ⅰでアップロード した" *fukushi.csv* " から、出現傾向を調べるのに必要な情報だけを抽出する。

1.    csvファイル" *fukushi.csv* " を、′￥ｔ’（タブ)で区切る。

2.   ５番目(キー)、７番目(語彙素読み)、８番目(語彙素)、１８番目(レジスター)のデータを取り出す

3.   ２の結果を表示して、 "kekka.csv" をダウンロードする

```
# -*-coding: utf-8 -*-

datafile = open('fukushi.csv')      #BCCWJコアデータから「副詞」を検索したfukushi.csvをファイルオープン
outputfile = open('kekka.csv','w')

for line in datafile:         #datafileから一行ずつ読み取る
  line = line.rstrip()        #行末の改行コードを取り除く
  columns = line.split('\t')  #「タブ」で区切る
  print(columns[4],",",columns[6],",",columns[7],",",columns[17],file=outputfile) #column[4](キー),columns[6](語彙素読み),columns[7](語彙素),columns[17](レジスタ―)を取り出して表示する
  
datafile.close()
outputfile.close()
files.download('kekka.csv')   #kekka.csvをダウンロードする
```

## Ⅳ．データ解析
###pandas を使ってデータフレームを作成し、データ解析をしやすくする。

1.    "pandas"　"numpy"　を 読み込む

2.    区切り記号 ' , '  で区切って、インデックス名（キー：'key' /語彙素読み：'lexyomi' /語彙素：'lexeme'  /レジスター：'register')を指定する

3.   'lexyomi'(語彙素読み)で　降順ソートする

4.   最初の10行を表示する

```
import pandas as pd      #panndasを使って、インデックス名を指定し、データフレームを作成する。
import numpy as np
import io
df = pd.read_csv(io.StringIO(uploaded['kekka.csv'].decode('utf-8')),delimiter=',',header=None,names=['key','lexyomi','lexeme','register'])
df.sort_values(by='lexyomi', ascending = False)   #lexyomi(語彙素読み)で降順ソートする
df.head(10)
```

##Ⅴ．データ集計
###ピボットテーブル機能を使って、出現頻度を計算する

1.   index（行）を 'lexeme'(語彙素)と 'key'(キー)に、'columns'（列）を 'register'（レジスター名)に指定して、出現頻度を計算する

2.  結果を表示して、 'fukushi_pivot.tsv' をダウンロードする




```
dfpivot= pd.pivot_table(df,index=['lexeme','key'],columns='register',aggfunc='count')  #indexをlexeme(語彙素),key(キー)に、columnsをregister(レジスター)に指定し、ピボットテーブル機能を使って、出現回数を集計する
dfpivot.head(100)

#dfpivot.to_csv('fukushi_pivot.tsv')
#files.download('fukushi_pivot.tsv')
```
